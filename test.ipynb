```python
# Create ExpenseManager
manager = ExpenseManager()

# Take user input
expense_id = int(input("Enter expense ID: "))
user_id = int(input("Enter your user ID: "))
amount = float(input("Enter amount: "))
category = input("Enter category (must exist in DB): ")
description = input("Enter description: ")

# Create Expense object
expense = Expense(expense_id, user_id, amount, category, description)

# Add to SQL
manager.add_expense(expense)

# Fetch all expenses to see result
all_expenses = manager.fetch_all_expenses()
print("All Expenses:")
for exp in all_expenses:
    print(exp)

# Fetch by category
cat_expenses = manager.fetch_expenses_by_category(category)
print(f"\nExpenses in category '{category}':")
for exp in cat_expenses:
    print(exp)
```
